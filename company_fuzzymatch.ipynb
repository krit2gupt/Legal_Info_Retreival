```python
import copy
final=[]

for c in choices.iteritems():
  s=c[1]
  l=s.split(" ")
  # print (l)
  l_new = []
  for i in l:
    temp = i.split("\t")
    for j in temp:
      l_new.append(j)
  l = copy.deepcopy(l_new)

  l_new = []
  for i in l:
    temp = i.split(".")
    for j in temp:
      l_new.append(j)
  l = copy.deepcopy(l_new)

  l_new = []
  for i in l:
    temp = i.split(")")
    for j in temp:
      l_new.append(j)
  l = copy.deepcopy(l_new)

  l_new = []
  for i in l:
    temp = i.split(",")
    for j in temp:
      l_new.append(j)
  l = copy.deepcopy(l_new)

  l_new = []
  for i in l:
    temp = i.split("(")
    for j in temp:
      l_new.append(j)
  l = copy.deepcopy(l_new)

  l_new = []
  for i in l:
    temp = i.split("-")
    for j in temp:
      l_new.append(j)
  l = copy.deepcopy(l_new)

  # print (l)
  l2=[]
  for x in l:
    x=x.lower()
    if x != "" and (not x.isnumeric()) and x not in ['m/s','m/s.','ms','the','branch','manager','branchmanager','private',
             'limited','through','pvt','ltd',
             'mr','office', 'co', 'anr', 'another',
             'rep','represented','its','lmt','managing','smt', 'sri', 'director']:
      l2.append(x)
  s=" ".join(l2)
  print(s)
  final.append(s)
comp_df=pd.DataFrame(final,columns=['company_name'])
```
